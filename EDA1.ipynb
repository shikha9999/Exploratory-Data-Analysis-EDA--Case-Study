<h2> Cumulative Distribution Funcation (CDF) </h2>

```python
import numpy as np

counts, bin_edges = np.histogram(haberman["age_of_patient"], bins=10, density =True)
pdf =counts/(sum(counts))
print(pdf);
print(bin_edges)
print("*"*80)
cdf= np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.grid(True)
plt.xlabel("Survival_status")
plt.ylabel("Age_of_patient")
plt.title('Cumulative Distribution Funcation Plot')
plt.show()
```

CDF of age of patient is easy to find survival status of patient.

```python
import numpy as np


counts, bin_edges = np.histogram(haberman["year_of_operation"], bins=10, density =True)
pdf =counts/(sum(counts))
print(pdf);
print(bin_edges)
print("*"*80)

cdf= np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.xlabel("Survival_status")
plt.ylabel("year_of_operation")
plt.title('Cumulative Distribution Funcation Plot')
plt.show()
```

 CDF of year of operation is easy to find survival status of patient

```python
import numpy as np

counts, bin_edges = np.histogram(haberman["nodes_detected"], bins=10, density =True)
pdf =counts/(sum(counts))
print(pdf);
print(bin_edges)
print("*"*80)
cdf= np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.xlabel("Survival_status")
plt.ylabel("Nodes_detected")
plt.title('Cumulative Distribution Funcation Plot')
plt.show()
```

CDF of nodes detected is easy to find survival status of patient.
